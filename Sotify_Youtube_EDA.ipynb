```python
import matplotlib.pyplot as plt

# Select numerical columns
numerical_columns = data.select_dtypes(include=['float64']).columns

# Plot histograms for each numerical feature
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(data[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.grid(True)
    plt.show()
```

```python

## Danceability ##
# Dans edilebilirliğin dağılımı sağa eğimlidir ve çoğu şarkı orta ile yüksek değerlere sahiptir.Dans etmeye uygun enerjik ve ritmik parçalara yönelik bir tercihi göstermektedir.

## Energy ##
# Enerji dağılımı geniş bir aralığa sahiptir ve daha yüksek seviyelerde zirveye ulaşır.Dinamik ve canlı müziğe olan bir tercihi yansıtır, ancak çeşitliliği gösteren daha düşük enerjili parçalar da vardır.

## Key ##
# 0 ve 1: Bu anahtarlar en yüksek frekanslara sahip.Bu tonların analiz edilen müziklerde en yaygın kullanılan anahtarlar olduğunu gösteriyor.
# 5: Bu anahtarın çok düşük frekanslı olması, bu tonda çok az müzik parçası olduğu anlamına gelir.

## Loudness ##
# Ses yüksekliği dağılımı oldukça yayılmış durumda, çoğu şarkı orta seviyelerde kümelenmiş durumda.Geniş bir kitleye rahat bir dinleme deneyimi sunan dengeli bir ses yüksekliği olduğunu gösteriyor.
    
## Speechiness ##
# Dağılım sağa eğimli görünüyor, çoğu şarkının düşük konuşma değerleri var.Şarkıların küçük bir kısmı daha yüksek konuşma değerleri sergiliyor, bu da konuşulan kelimelerin varlığını gösteriyor.

## Acousticness ##
# Dağılım nispeten eşit bir şekilde yayılmış gibi görünüyor, daha düşük değerler civarında bir tepe noktası var.Daha yüksek akustikliğe sahip şarkıların belirgin bir oranı var, bu da önemli bir akustik bileşen olduğunu gösteriyor.

## Instrumentalness ##
# Dağılım düşük değerlere doğru oldukça eğimlidir, bu da çoğu şarkının vokal içerdiği anlamına gelir.

## Liveness ##
# Dağılım biraz sağa eğimli görünüyor, şarkıların çoğunluğu daha düşük canlılık değerlerine sahip.Daha az şarkı daha yüksek canlılık sergiliyor, bu da daha az oranda canlı kayıt olduğunu gösteriyor.

## Valence ##
# Dağılım nispeten simetrik görünüyor.Şarkıların hem olumlu hem de olumsuz duygusal tonları iletmesiyle değerde değişkenlik var.

## Tempo ##
# Dağılım yaklaşık olarak normal görünüyor, orta tempo etrafında merkezlenmiş.Bazı uç değerler çok düşük veya çok yüksek tempo değerlerine sahip, ancak çoğu şarkı orta tempo aralığına giriyor.

## Duration_ms ##
# Dağılım sağa çarpıktır ve tepe noktası daha kısa sürelerdedir.Bazı şarkılar çok daha uzun sürelere sahiptir ve bu da parça uzunluğundaki değişkenliği gösterir.
```

```python
columns_to_plot = [
    'Speechiness', 'Acousticness', 'Instrumentalness', 
    'Liveness', 'Valence', 'Tempo', 'Duration_ms', 
    'Views', 'Likes', 'Comments', 'Stream'
]
data_selected = data[columns_to_plot]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 40))
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(len(columns_to_plot), 1, i)
    sns.boxplot(data=data[column], orient="v")  
    plt.title(f"Distribution of {column}")
    plt.ylabel("Value")

plt.tight_layout()
plt.show()
```
